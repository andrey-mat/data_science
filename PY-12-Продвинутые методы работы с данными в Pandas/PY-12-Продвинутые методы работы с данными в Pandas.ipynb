```python
import pandas as pd
melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
melb_df['quarter'] = melb_df['Date'].dt.quarter
melb_df['quarter'].value_counts().iloc[1]
```

```python
melb_df.drop(['quarter'], axis=1, inplace=True)
```

```python
count = 0
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
for col in melb_df.columns:
    if melb_df[col].nunique() < 150 and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')
        count += 1
count
```

```python
melb_df.sort_values(by=['Distance', 'Price']).loc[::50,['Distance', 'Price']]
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_df.sort_values(by=['AreaRatio'],
                    ascending=False,
                    ignore_index=True).loc[1558]['BuildingArea']
```

```python
mask1 = melb_df['Type'] == 'townhouse'
mask2 = melb_df['Rooms'] > 2
melb_df[mask1 & mask2].sort_values(
    by=['Rooms', 'MeanRoomsSquare'],
    ascending=[True, False],
    ignore_index=True).loc[18]['Price']
```

```python
melb_df.groupby('Regionname')['Distance'].mean().sort_values(ascending=False)
```

```python
melb_df.groupby('MonthSale')['Price'].agg(
    'describe'
    ).sort_values(by='count', ascending=False)
```

```python
melb_df.groupby('Regionname')['SellerG'].agg(
    ['nunique', set]
)
```

```python
date1 = pd.to_datetime('2017-05-01')
date2 = pd.to_datetime('2017-09-01')
mask = (melb_df['Date'] > date1) & (melb_df['Date'] < date2)
melb_df.loc[mask].groupby('SellerG')['Price'].sum().sort_values()
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].median().unstack()
```

```python
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
mask = pivot['mean']['house'] < pivot['median']['house']
filtered_pivot = pivot[mask]
filtered_pivot.index
```

```python
pivot = melb_df.pivot_table(
    values='BuildingArea',
    aggfunc='median',
    columns='Type',
    index='Rooms',
    fill_value=0
)
```

```python
pivot = melb_df.pivot_table(
    values='Price',
    aggfunc='median',
    index='SellerG',
    columns='Type',
    fill_value=0
)
pivot['unit'].sort_values(ascending=False).index[0]
```
