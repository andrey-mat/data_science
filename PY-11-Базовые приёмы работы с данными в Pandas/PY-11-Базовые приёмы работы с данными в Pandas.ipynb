```python
import pandas as pd

melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
melb_data.head()
```

```python
melb_df = melb_data.copy()
melb_df.head()
```

```python
melb_df.drop(['index','Coordinates'],axis=1,inplace=True)
melb_df.head()
```

```python
total_rooms = melb_df['Rooms'] + melb_df['Bedroom'] + melb_df['Bathroom']
melb_df['MeanRoomsSquare'] = melb_df['BuildingArea'] / total_rooms
melb_df.head()
```

```python

diff_area = melb_df['BuildingArea'] - melb_df['Landsize']
sum_area = melb_df['BuildingArea'] + melb_df['Landsize']
melb_df['AreaRatio'] = diff_area / sum_area
melb_df.loc[:, ['BuildingArea', 'Landsize', 'AreaRatio']]
```

```python
import pandas as pd
def delete_columns(df, col=[]):
    try:
        df = df.drop(col, axis=1)
        return df
    except KeyError:
        return None

customer_df = pd.DataFrame({
        'number': [0, 1, 2, 3, 4],
        'cust_id': [128, 1201, 9832, 4392, 7472],
        'cust_age': [13, 21, 19, 21, 60],
        'cust_sale': [0, 0, 0.2, 0.15, 0.3],
        'cust_year_birth': [2008, 2000, 2002, 2000, 1961],
        'cust_order': [1400, 14142, 900, 1240, 8430]
    })
customer_df = delete_columns(customer_df, ['cust_i', 'cust_age'])
display(customer_df)

```

```python
melb_df['Address'].nunique()

def get_street_type(address):
    exclude_list = ['N', 'S', 'W', 'E']
    short_streets = {'Av': 'Avenue', 'Bvd': 'Boulevard', 'Pde': 'Parade'}
    address = address.split()
    result = address[-1]
    if result in exclude_list:
        result = address[-2]
    if result in short_streets:
        result = short_streets[result]
    return result


street_types = melb_df['Address'].apply(get_street_type)
popular_types = street_types.value_counts().nlargest(10).index
melb_df['StreetType'] = street_types.apply(lambda x:x if x in popular_types else 'other')
melb_df['StreetType'].value_counts(normalize=True)
```

```python
melb_df.drop('Address', axis=1, inplace=True)
melb_df
```

```python
def get_weekend(weekday):
    if weekday >= 5:
        return 1
    else:
        return 0

melb_df['Weekend'] = melb_df['WeekDaySale'].apply(get_weekend)
round(melb_df[melb_df['Weekend'] == 1]['Price'].mean())
```

```python
best_sales = melb_df['SellerG'].value_counts().nlargest(49).index
melb_df['SellerG'] = melb_df['SellerG'].apply(lambda s: s if s in best_sales else 'other')
melb_df['SellerG'].value_counts(normalize=True)
mask1 = melb_df['SellerG'] == 'Nelson'
mask2 = melb_df['SellerG'] == 'other'
round(melb_df[mask1]['Price'].min() / melb_df[mask2]['Price'].min(), 1)
```

```python
unique_list = []
for col in melb_df.columns:
    item = (col, melb_df[col].nunique(), melb_df[col].dtype)
    unique_list.append(item)

unique_counts = pd.DataFrame(
    unique_list,
    columns=['Column_name', 'Num_Unique','Type']
).sort_values(by='Num_Unique', ignore_index=True)

display(melb_df.info())
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_unique_counts = 150
for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_counts and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')
melb_df.info()
```

```python
melb_df['Type'] = melb_df['Type'].cat.rename_categories({
    'u': 'unit',
    't': 'townhouse',
    'h': 'house'
})
display(melb_df['Type'])
```

```python
new_houses_types = pd.Series(['unit', 'house', 'flat', 'flat', 'house'])
new_houses_types = new_houses_types.astype(melb_df['Type'].dtype)
display(new_houses_types)
melb_df['Type'] = melb_df['Type'].cat.add_categories('flat')
new_houses_types = pd.Series(['unit', 'house', 'flat', 'flat', 'house'])
new_houses_types = new_houses_types.astype(melb_df['Type'].dtype)
display(new_houses_types)
```
